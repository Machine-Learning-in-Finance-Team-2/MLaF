```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import time
```

```python
def get_news_cnntech(year, month):
    # url definition
    url = "https://www.cnn.com/tech/article/sitemap-" + str(year) + "-" + str(month) +".html"

    # Request
    r1 = requests.get(url)
    r1.status_code

    # We'll save in coverpage the cover page content
    coverpage = r1.content

    # Soup
    soup1 = BeautifulSoup(coverpage, "html5lib")

    # News identification
    coverpage_news = soup1.find_all('span', class_="sitemap-link")
    coverpage_date = soup1.find_all('span', class_="date")

    number_of_articles = len(coverpage_news)

    # Empty lists for content, links and titles
    list_titles = []
    list_dates = []

    for n in np.arange(1, number_of_articles):

        # Getting the title
        title = coverpage_news[n].find('a').get_text()
        list_titles.append(title)

        # Getting the date
        date = coverpage_date[n].get_text()
        list_dates.append(date)

        # df_show_info
        df_show_info = pd.DataFrame(
        {'Article Title': list_titles,
        'Article Date': list_dates})

    return (df_show_info)
```

```python
get_news_cnntech(2012,8)
```

```python
list_titles = []
list_dates = []

final_df = pd.DataFrame(
{'Article Title': list_titles,
'Article Date': list_dates})

#append info of each coverpage to final dataframe; theguardian has 1900 pages
for i in range(2012,2017):
    for j in range(1,13):
        final_df = final_df.append(get_news_cnntech(i,j))
```
