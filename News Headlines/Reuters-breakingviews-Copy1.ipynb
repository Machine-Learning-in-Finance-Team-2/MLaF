```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import time
```

```python
def get_news_reuters(page):
    # url definition
    url = "https://www.reuters.com/news/archive/mcbreakingnews?view=page&page=" + str(page) + "&pageSize=10"

    # Request
    r1 = requests.get(url)
    r1.status_code

    # We'll save in coverpage the cover page content
    coverpage = r1.content

    # Soup
    soup1 = BeautifulSoup(coverpage, "html5lib")

    # News identification
    coverpage_news = soup1.find_all('h3', class_="story-title")
    coverpage_date = soup1.find_all('time')

    # Empty lists for content, links and titles
    list_titles = []
    list_dates = []

    for n in np.arange(0, 10):

        # Getting the title
        title = coverpage_news[n].get_text()
        list_titles.append(title.replace('\t', '').replace('\n',''))

        # Getting the date
        date = coverpage_date[n].find('span').get_text()
        list_dates.append(date)

        # df_show_info
        df_show_info = pd.DataFrame(
        {'Article Title': list_titles,
        'Article Date': list_dates})

    return (df_show_info)
```

```python
get_news_reuters(100)
```

```python
list_titles = []
list_dates = []

final_df = pd.DataFrame(
{'Article Title': list_titles,
'Article Date': list_dates})

#append info of each coverpage to final dataframe; theguardian has 1900 pages
for i in range(1,3277):
    final_df = final_df.append(get_news_reuters(i))
```
